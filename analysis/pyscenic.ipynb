```python
import json
import zlib
import base64

# collect SCENIC AUCell output
lf = lp.connect("analysis/pyscenic/pyscenic_output.loom", mode='r+', validate=False )
auc_mtx = pd.DataFrame( lf.ca.RegulonsAUC, index=lf.ca.CellID)
lf.close()
```

```python
import umap

# UMAP
runUmap = umap.UMAP(n_neighbors=10, min_dist=0.4, metric='correlation').fit_transform
dr_umap = runUmap(auc_mtx)
pd.DataFrame(dr_umap, columns=['X', 'Y'], index=auc_mtx.index).to_csv("analysis/pyscenic/scenic_umap.txt", sep='\t')
```
